```python
import cv2
import matplotlib.pyplot as plt

def convert_to_bgr(image_path, output_path):
    # Load the image (in OpenCV, this loads as BGR by default)
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Image not found or path is incorrect.")
        return

    # Convert image from RGB to BGR (though it's already BGR from imread)
    bgr_image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Display the original image
    print("Original Image:")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
    plt.axis('off')
    plt.show()

    # Display the BGR image (will look different if interpreted as RGB)
    print("BGR Image:")
    plt.imshow(bgr_image)  # This will show wrong colors since it's BGR
    plt.axis('off')
    plt.show()

    # Save the converted image
    cv2.imwrite(output_path, bgr_image)
    print("BGR image saved as", output_path)



```

```python
import cv2

def convert_bgr_to_rgb(image_path):
    # Read image in BGR format (default for OpenCV)
    bgr_img = cv2.imread(image_path)

    if bgr_img is None:
        print("Image not found.")
        return

    # Convert to RGB
    rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)

    # Save the RGB image (convert back to BGR for saving with OpenCV)
    output_path_rgb = image_path.replace(".jpg", "_rgb.jpg")
    cv2.imwrite(output_path_rgb, cv2.cvtColor(rgb_img, cv2.COLOR_RGB2BGR))

    print("Image converted to RGB and saved at:", output_path_rgb)

# Your existing line
convert_bgr_to_rgb(output_path)
```

```python
def create_white_image(width, height):
  white_image=np.ones((height,width,3), dtype=np.uint8)*255

  print("White Image:")
  cv2.imshow("White Image",white_image)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```

```python
def create_black_image(width, height):
    try:
        # Create a black color image
        black_image = np.zeros((height, width, 3), dtype=np.uint8)

        # Display the black image
        cv2.imshow("Black Image", black_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
import cv2

def draw_shapes_on_image(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Resize image for consistent display (optional)
    scale_width = 800
    scale_factor = scale_width / image.shape[1]
    new_dim = (scale_width, int(image.shape[0] * scale_factor))
    image = cv2.resize(image, new_dim)

    # Draw a red rectangle around the face
    cv2.rectangle(image, (240, 150), (380, 300), (0, 0, 255), 2)

    # Draw a green circle around the trophy
    cv2.circle(image, (520, 240), 100, (0, 255, 0), 2)

    # Draw a blue line from mouth to trophy
    cv2.line(image, (360, 230), (480, 230), (255, 0, 0), 2)

    # Display the image
    cv2.imshow("Image with Shapes", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

```
